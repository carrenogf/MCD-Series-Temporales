<h3>Pregunta 1= Describa las caracteristicas principales de un modelo autoregresivo</h3>

1. Dependencia de valores pasados: Un modelo AR utiliza los valores anteriores de la serie temporal para predecir los futuros. Esto significa que los valores futuros son una funci贸n lineal de los valores anteriores y un t茅rmino de error aleatorio.

2. Orden del modelo: El orden de un modelo AR, denotado como AR(p), indica el n煤mero de valores pasados utilizados en la predicci贸n. Por ejemplo, un modelo AR(1) usa el valor inmediatamente anterior, mientras que un AR(2) usa los dos valores anteriores.
3. Estacionariedad: Los modelos AR son m谩s efectivos con series temporales estacionarias, lo que significa que las propiedades estad铆sticas (como la media y la varianza) son constantes en el tiempo.

<h3>Pregunta 2= Describa las caracteristicas principales de un modelo de medias moviles</h3>

1. Dependencia de errores pasados: A diferencia del modelo AR, un modelo MA utiliza los errores (residuos) pasados para predecir el valor futuro. Estos errores representan la parte impredecible de la serie.

2. Orden del modelo: El modelo MA(q) tiene un orden q que indica el n煤mero de errores pasados utilizados en la predicci贸n. Por ejemplo, un MA(1) usa el error del periodo anterior, mientras que un MA(2) usa los errores de los dos periodos anteriores.
3. Estacionariedad: Los modelos MA son siempre estacionarios, ya que no dependen de los valores pasados de la serie, sino de los errores, que se asumen como ruido blanco.

<h3>Pregunta 4= 驴 cual es el criterio utilizado mas destacado para determinar el orden optimo de un modelo AR, MA o ARMA.?</h3>

El criterio m谩s destacado y ampliamente utilizado para determinar el orden 贸ptimo de un modelo AR, MA, o ARMA es el Criterio de Informaci贸n de Akaike (AIC). Sin embargo, tambi茅n se utilizan otros criterios como el Criterio de Informaci贸n Bayesiano (BIC) y el Criterio de Hannan-Quinn (HQIC). A continuaci贸n, te explico los m谩s importantes:


<b>Criterio de Informaci贸n de Akaike (AIC):</b> <br>
Descripci贸n: El AIC es una medida que eval煤a la calidad de un modelo en funci贸n de su capacidad de ajuste (verosimilitud) y su simplicidad (penaliza el n煤mero de par谩metros). El objetivo es encontrar el modelo con el menor valor de AIC.

<b>Criterio de Informaci贸n Bayesiano (BIC):</b> <br>
Descripci贸n: El BIC es similar al AIC, pero penaliza de manera m谩s fuerte el n煤mero de par谩metros. Esto significa que BIC tiende a seleccionar modelos m谩s simples en comparaci贸n con AIC.

<b>Criterio de Hannan-Quinn (HQIC) </b><br>
Descripci贸n: Es otro criterio similar a AIC y BIC, pero tiene una penalizaci贸n intermedia, menos fuerte que BIC y m谩s fuerte que AIC.

<b>Comparaci贸n y Elecci贸n:</b><br>
* AIC tiende a seleccionar modelos m谩s complejos (mayor n煤mero de par谩metros), ya que penaliza de manera m谩s suave.
* BIC es m谩s conservador, prefiriendo modelos m谩s simples, por lo que es m谩s adecuado cuando se tiene un conjunto de datos grandes y se quiere evitar el sobreajuste.
* HQIC es una alternativa que penaliza de manera m谩s fuerte que AIC, pero menos que BIC, y a veces puede ser 煤til en casos donde ni AIC ni BIC funcionan bien.

En resumen, el AIC es el criterio m谩s destacado y com煤nmente utilizado, pero en contextos donde se desea un mayor control sobre la complejidad del modelo, el BIC tambi茅n es una excelente opci贸n.

<h3>驴Como determinar el orden p, d, q de un modelo AR, MA y/o ARMA?</h3>

Para determinar estos 贸rdenes, se siguen generalmente estos pasos:

1. An谩lisis gr谩fico:
  * Gr谩fico de la serie temporal
  * Funci贸n de Autocorrelaci贸n (ACF)
  * Funci贸n de Autocorrelaci贸n Parcial (PACF)


2. Criterios de informaci贸n:
  * AIC (Criterio de Informaci贸n de Akaike)
  * BIC (Criterio de Informaci贸n Bayesiano)


3. Pruebas de diagn贸stico:
  * An谩lisis de residuos
  * Pruebas de normalidad
  * Pruebas de autocorrelaci贸n de residuos



El proceso t铆pico es:

1. Determinar d: Se analiza la estacionariedad de la serie. Si no es estacionaria, se diferencia hasta lograr la estacionariedad.

2. Determinar p y q:
  * Para AR: Se observa la PACF. El n煤mero de rezagos significativos sugiere el orden p.
  * Para MA: Se observa la ACF. El n煤mero de rezagos significativos sugiere el orden q.
  * Para ARMA: Se comparan varios modelos con diferentes combinaciones de p y q.


3. Selecci贸n del modelo:

  * Se ajustan varios modelos con diferentes 贸rdenes.
  * Se comparan utilizando AIC o BIC. El modelo con el valor m谩s bajo de estos criterios suele considerarse el mejor.


3. Validaci贸n del modelo:

  * Se realizan pruebas de diagn贸stico para asegurar que el modelo es adecuado.

<h3>Pregunta 5= 驴Cu谩l es la raz贸n para optar por el principio de la parsimonia en la toma de decisiones para la seleccion de modelos ? </h3>

El principio de parsimonia, tambi茅n conocido como la navaja de Occam, es fundamental en la selecci贸n de modelos estad铆sticos, incluidos los modelos AR, MA y ARMA. La raz贸n principal para optar por este principio en el contexto de estos modelos se basa en varios aspectos importantes:

1. Evitar el sobreajuste:
Un modelo m谩s complejo (con m谩s par谩metros) tiende a ajustarse mejor a los datos de entrenamiento, pero puede "aprender" el ruido en los datos en lugar de la verdadera estructura subyacente. Esto puede llevar a un rendimiento deficiente en nuevos datos.

2. Mejor generalizaci贸n:
Los modelos m谩s simples suelen generalizar mejor a nuevos datos, lo que es crucial para hacer predicciones precisas.
3. Interpretabilidad:
Los modelos m谩s simples son generalmente m谩s f谩ciles de interpretar y explicar, lo que es importante en muchos contextos de toma de decisiones.
4. Estabilidad:
Los modelos m谩s simples tienden a ser m谩s estables y menos sensibles a peque帽os cambios en los datos de entrada.
5. Eficiencia computacional:
Modelos m谩s simples requieren menos recursos computacionales para ser estimados y utilizados.
6. Reducci贸n del error de estimaci贸n:
Cada par谩metro adicional introduce un error de estimaci贸n. Al minimizar el n煤mero de par谩metros, se reduce este error acumulativo.
7. Balance entre sesgo y varianza:
Los modelos m谩s simples pueden tener un mayor sesgo pero menor varianza, lo que a menudo resulta en un mejor rendimiento general.
8. Criterios de informaci贸n:
Los criterios como AIC y BIC incluyen penalizaciones por la complejidad del modelo, favoreciendo as铆 modelos m谩s simples.

En el contexto espec铆fico de los modelos AR, MA y ARMA:

* Para un modelo AR, preferir un orden p m谩s bajo significa asumir que los valores pasados m谩s recientes son los m谩s influyentes.
* Para un modelo MA, un orden q m谩s bajo implica que solo los errores m谩s recientes son significativos para predecir el valor actual.
* En modelos ARMA, buscar la combinaci贸n m谩s simple de p y q que capture adecuadamente la din谩mica de la serie temporal.

Es importante notar que la parsimonia no significa siempre elegir el modelo m谩s simple, sino encontrar un equilibrio entre la simplicidad y la capacidad explicativa. El objetivo es capturar la estructura esencial de los datos sin incluir complejidad innecesaria.

<h3>Pregunta 6= 驴Cu谩les son las dos propiedades clave que deben tener los residuos para que un m茅todo de pron贸stico se considere bueno seg煤n el texto? </h3> - Puede tomar como referencia el capitulo 3 del libro recomendado de Hyndman.

Bas谩ndome en el cap铆tulo 3 del libro de Hyndman sobre pron贸sticos, las dos propiedades clave que deben tener los residuos para que un m茅todo de pron贸stico se considere bueno son:

1. Los residuos deben ser no correlacionados.
2. Los residuos deben tener media cero.

Estas propiedades son fundamentales por las siguientes razones:

1. Residuos no correlacionados:

  * Significa que los residuos no deben mostrar ning煤n patr贸n.
  * Si hay correlaci贸n en los residuos, indica que hay informaci贸n en los errores que deber铆a ser utilizada en el pron贸stico.
  * La ausencia de correlaci贸n sugiere que toda la informaci贸n 煤til ha sido extra铆da de los datos.


2. Residuos con media cero:

  * Implica que los pron贸sticos no est谩n sesgados.
  * Si los residuos tienen una media diferente de cero, los pron贸sticos est谩n sistem谩ticamente muy altos o muy bajos.



Estas propiedades se pueden verificar mediante:

* Gr谩ficos de los residuos a lo largo del tiempo para detectar patrones.
* Gr谩ficos de autocorrelaci贸n (ACF) de los residuos para comprobar la correlaci贸n.
* Pruebas estad铆sticas como el test de Ljung-Box para la autocorrelaci贸n.
* C谩lculo de la media de los residuos y pruebas t para verificar si es significativamente diferente de cero.

Es importante notar que, aunque Hyndman destaca estas dos propiedades como las m谩s importantes, en la pr谩ctica tambi茅n se suelen considerar otras caracter铆sticas de los residuos, como la normalidad y la homocedasticidad (varianza constante), especialmente para ciertos tipos de modelos y pruebas estad铆sticas.

```python
# Aqui hacemos el loop probando diferentes "p" y "q", para saber cual es el criterio de informacion mas bajo.

import itertools
import statsmodels.api as sm

p_values = range(0, 3)
q_values = range(0, 3)


resultados = []
mejor_aic = float("inf")
mejor_order = None
mejor_modelo = None


for p, q in itertools.product(p_values, q_values):
    try:

        modelo = sm.tsa.arima.ARIMA(df, order=(p, 0, q))
        resultado = modelo.fit()

        aic = resultado.aic
        resultados.append((p, q, aic))

        if aic < mejor_aic:
            mejor_aic = aic
            mejor_order = (p, q)
            mejor_modelo = resultado

    except Exception as e:
        print(f"Error en el modelo ARMA({p},{q}): {e}")

print(f"Mejor modelo ARMA: order={mejor_order} con AIC={mejor_aic}")

for p, q, aic in resultados:
    print(f"ARMA({p},{q}) AIC: {aic}")
```
